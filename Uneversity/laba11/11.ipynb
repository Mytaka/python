Завдання 1: Заданий файл dict.txt, що містить словник в форматі:
cat - кішка
dog - собака
house - дім
В словнику англійські та українські слова розділені двома табуляціями та девісом: '\t-\t'.
У файлі translate.txt міститься текст для перекладу.
Необхідно зробити порядковий переклад та за допомогою словника вивести результат у файл output.txt. Незнайомі слова необхідно залишити у незмінному вигляді. 


```python
def to_dict(arr):
    return {arr[0]:arr[1]}


with open('dict.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

dict = {}

for line in lines[:-1]:
    line = line.strip().split('\t-\t')
    dict.update(to_dict(line))


with open('translate.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

words = []

for line in lines:
    line = line.strip()
    words += line.split(' ')


translate = [dict[word] if word in dict.keys() else word for word in words]
    
translate = [word+'\n' if word.endswith('.') else word+' ' for word in translate]

translate[3] = translate[3] + '\n\n'


with open('output.txt', 'w', encoding='utf-8') as file_end:
    file_end.write(''.join(translate))


print(dict)
print(words)
print(translate)
```

Завдання 2:
У файлі input_countries.txt заданий перелік країн та мов, на яких розмовляють в цих країнах у форматі: <Назва країни> : <мова 1> <мова 2> <мова 3>
Для введеної користувачем мови необхідно вивести перелік країн, в яких розмовляють даною мовою.


```python
def to_dict(arr):
    return {arr[0]:arr[1:]}

user =  'німецька'

with open('input_countries.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()

dict = {}

for line in lines:
    dict.update(to_dict(line.strip().split(' ')))

for key, value in dict.items():
    if user in value:
        print(key[:-1])
```

Завдання 3: У вагоні-купе є кілька купе, у кожному з яких по 4 місця. Розробник зберігає інформацію про зайнятість одного купе у вигляді словника:
{1: 'м', 2: None, 3: None, 4: 'ж'}

Визначте:
·        Кількість повністю вільних купе;
·        Кількість вільних місць у вагоні;
·        Кількість вільних нижніх чи верхніх місць;
·        Кількість вільних місць у поєднанні з виключно чоловічою компанією;
·        Кількість вільних місць у поєднанні з виключно жіночою компанією.


```python
from numpy import random
random.seed(42)

possibilities = [None,'м','ж']

# 1 вагон-купе в якому 200 купе і в кожному купе по 4 місця
dict = { i : { j : random.choice( possibilities ) for j in range(1,5) } for i in range(1,200) }

count_none = 0
count_free_coupe = 0
count_upper_seat, count_down_seat = 0,0
count_seat_w_men = 0
count_seat_w_women = 0

for coupe in dict.values():
    
    is_men = 0
    is_women = 0

    for seat, status in coupe.items():

        if status is None: 

            is_men += 1
            is_women += 1
            count_none += 1

            if seat in [1, 2]:  
                count_down_seat += 1

            elif seat in [3, 4]:  
                count_upper_seat += 1

        elif status == 'м':
            is_men += 1
            
        elif status == 'ж':
            is_women += 1
        
    if is_men == 4:
        count_seat_w_men += 1

    elif is_women == 4:
        count_seat_w_women += 1
    
print('Кількість повністю вільних купе: ', count_free_coupe)
print('Кількість вільних місць у вагоні: ', count_none)
print('Кількість вільних нижніх: ', count_down_seat, '\nверхніх місць: ', count_upper_seat)
print('Кількість вільних місць у поєднанні з виключно чоловічою компанією: ', count_seat_w_men)
print('Кількість вільних місць у поєднанні з виключно жіночою компанією: ',count_seat_w_women)
print(dict)
```
