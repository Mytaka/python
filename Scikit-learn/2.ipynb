```python
import matplotlib.pyplot as plt

year = ['Первый', 'Второй', 'Третий', 'Четвертый', 'Пятый']
frequency = [500, 470, 300, 260, 220]

plt.bar(year,frequency)
plt.xlabel('Курс')
plt.ylabel('Количество студентов')
```

```python
import numpy as np
np.random.seed(42)

height = np.round(np.random.normal(180,10,1000))
print(height)
```

```python
bins = 10
```

```python
plt.hist(height,bins)
plt.xlabel('Рост респондентов, см') 
plt.ylabel('Количество респондентов') 
plt.title('Распределение роста респондентов-мужчин в Украине')
```

```python
import statistics as stat

stat.mean(height)
stat.median(height)
stat.mode(height)
stat.stdev(height)
```
