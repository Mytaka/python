```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42) 
height_men = list(np.round(np.random.normal(180, 10, 1000))) 
height_women = list(np.round(np.random.normal(160, 10, 1000)))
```

```python
bins = 15

plt.figure(figsize=(10,6)) # задаем размер

plt.hist(height_men, bins, alpha=0.5, label='Рост мужчин')   # alpha - прозрачность
plt.hist(height_women, bins, alpha=0.5, label='Рост женщин')

plt.legend(loc='upper right')
plt.xlabel('Рост, см', fontsize=16) 
plt.ylabel("Количество респондентов", fontsize=16) 
plt.title('Pacпределение роста мужчин и женщин в Украине', fontsize=18)
```

```python
height_women_new = [1.48, 1.49, 1.49, 1.50, 1.51, 1.52, 1.52, 1.53, 1.53, 1.54, 1.55, 1.56, 1.57, 1.57, 1.58, 1.58, 1.59, 1.60, 1.61,1.62]
neck_women = [29.1, 30.0, 30.1, 30.2, 30.4, 30.6, 30.8, 30.9, 31.0, 30.6, 30.7, 30.9, 31.0, 31.2, 31.3, 32.0, 31.4, 31.9, 32.4, 32.8]

plt.figure(figsize=(10,6))
plt.scatter(height_women_new,neck_women) # точечный гравфик

plt.xlabel('Pост, м', fontsize=16)
plt.ylabel('0бхват шеи, см', fontsize=16)
plt.title('Зависимость роста и окружности шеи у женщин в Украине', fontsize=18)
```

```python
import scipy.stats as st
print( round(st.pearsonr(height_women_new,neck_women)[0], 2) ) # кореляция 
np.corrcoef(height_women_new,neck_women)                       # кореляция 
```

```python
plt.figure(figsize=(10,6))

x = np.array(height_women_new)
y = np.array(neck_women)

plt.scatter(x,y)

slope, intercept = np.polyfit(x, y, 1)

plt.plot(x, x*slope + intercept, 'r')

plt.xlabel('Pост, м', fontsize=16)
plt.ylabel('0бхват шеи, см', fontsize=16)
plt.title('Зависимость роста и окружности шеи у женщин в Украине', fontsize=18)
```

```python
ci = st.norm.interval(confidence=0.90, loc=np.mean(height_women_new), scale=st.sem(height_women_new))

print((round(ci[0], 2), round(ci[1], 2)))
```
